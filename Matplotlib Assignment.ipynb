1. Create a pie chart presenting the male/female proportion

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
titanic.head()
```

```python

titanic.sex.value_counts().plot.pie(figsize=(8,8), fontsize=20,autopct='%.2f')
                                 
```

```python
#Method 2
titanic.groupby('sex').size().plot.pie(autopct='%.2f')
```
